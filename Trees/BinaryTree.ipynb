```python
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
btn1 = BinaryTreeNode(3)
btn2 = BinaryTreeNode(7)
btn3 = BinaryTreeNode(6)
btn4 = BinaryTreeNode(5)
btn5 = BinaryTreeNode(4)
btn1.left = btn2
btn1.right = btn3
btn2.left = btn4
btn2.right = btn5
print(btn1.data)
print(btn1.left.data)
```

```python
def printTree(root):
    if root is None:
        return
    print(root.data)
    printTree(root.left)
    printTree(root.right)
```

```python
printTree(btn1)
```

```python
def inputTree():
    rootData = int(input())
    if rootData == -1:
        return None

    root = BinaryTreeNode(rootData)
    leftTree = inputTree()
    rightTree = inputTree()
    root.left = leftTree
    root.right = rightTree

    return root
    
    
```

```python
root = inputTree()
printTree(root)
```
