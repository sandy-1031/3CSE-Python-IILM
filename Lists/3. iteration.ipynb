```python
for i in range(len(li)):
    print(li[i])
```

```python
# input 1 = size of list
# input 2 = elements seperated by space
# prints the sum of these elements
### ignore strip() for now

#n = int(input())
arr = list(int(i) for i in input().strip().split(' '))
sum = 0
for number in arr:
    sum += number
print(sum)
```
