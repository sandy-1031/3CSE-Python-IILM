Given an array of integers arr[], the task is to find the maximum and minimum elements in the array using the minimum number of comparisons.

```python
def findArr(arr):
    if len(arr) == 0:
        return 
    #return (max(arr), min(arr))
    maxele = -1
    minele = 100000000000
    for i in range(len(arr)):
        if arr[i] > maxele:
            maxele = arr[i]
        elif arr[i] < minele:
            minele = arr[i]
    
    return(maxele, minele)
    

arr = [1,7,2,4,10,12,15, 4, 3]
ans = findArr(arr)
print(ans)
```

Reverse the Array: Reverse an array arr[]. Reversing an array means rearranging the elements such that the first element becomes the last, the second element becomes second last and so on.

```python
def reverseArr(arr):
    if len(arr) <= 1:
        return arr
    return arr[::-1]

arr = [1,7,2,4,10,12,15, 4, 3]
ans = reverseArr(arr)
print(ans)
```

Kth Smallest: Given an integer array arr[] and an integer k, your task is to find and return the kth smallest element in the given array.

```python
def k_small(arr, k):
    arr.sort()
    return arr[k-1]

arr = [1,7,2,4,10,12,15, 4, 3]
k = 3
ans = k_small(arr, k)
print(ans)
```

Sort 0s, 1s and 2s: Given an array arr[] containing only 0s, 1s, and 2s. Sort the array in ascending order. Note: You need to solve this problem without utilizing the built-in sort function. (AKA Dutch National Flag Algorithim)

```python
def sort012(arr):
    start, mid = 0, 0
    end = len(arr) - 1

    while mid <= end:
        if arr[mid] == 0:
            arr[start], arr[mid] = arr[mid], arr[start]
            start += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:  # the value is 2 for mid
            arr[mid], arr[end] = arr[end], arr[mid]
            end -= 1
    return arr

arr = [0,0,1,2,1,2,0,2,1,0,2,1,2,1,0]
ans = sort012(arr)
print(ans)
```

Move all negative numbers to beginning and positive to end with constant extra space.
Input: -12, 11, -13, -5, 6, -7, 5, -3, -6 
Output: -12 -13 -5 -7 -3 -6 11 6 5


```python
def movenums(arr):
    if len(arr) == 0:
        return arr
    start = 0
    end = len(arr)-1
    while start <= end:
        if arr[start] < 0:
            start += 1
        elif arr[end] > 0:
            end -= 1
        else:
            arr[start], arr[end] = arr[end], arr[start]

    return arr

ans = movenums([-12, 11, -13, -5, 6, -7, 5, -3, -6 ])
print(ans)
```
