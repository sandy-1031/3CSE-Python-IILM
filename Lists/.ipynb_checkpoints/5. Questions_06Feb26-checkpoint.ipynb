```python
#Union Arrays

arr1 = [1,2,1,2,3,4]
arr2 = [3,3,4,5,6]
#Union of array = [1,2,3,4,5,6]

#Approach - storing the occurance of elements in dictionary
d = {}
#Traversing the first Array
for ele in arr1:
    if ele in d:
        d[ele] += 1
    else:
        d[ele] = 1
#Traversing the second array
for ele in arr2:
    if ele in d:
        d[ele] += 1
    else:
        d[ele] = 1
ans = []
for key in d:
    ans.append(key)

print("The union of arr1 and arr2 is: ", ans)
```

```python
#Kadane's Algo

def kadanesAlgo(arr):   #Return max sum of subarray
    res = arr[0]
    maxSum = arr[0]

    for i in range(1, len(arr)):
        maxSum = max(maxSum + arr[i], arr[i])
        res = max(maxSum, res)
    return res

arr = [2, 3, -2, -1, -1, -1, 8, 9]
print(kadanesAlgo(arr))
```
