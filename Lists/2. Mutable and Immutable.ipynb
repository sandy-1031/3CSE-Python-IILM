Immutable Concept - the value is not changed after implementing the changes but the reference is changed with respect to variable.
Mutable concept - it is applied in case of lists, when 2 lists are pointing to same reference list, and changes are done at the list. Then the changes will be reflected on both of the variables which were pointing to the same list 

```python
li = [1,2,3,4]
li2 = li
#li2 = [3,3,4] #reference is changed
li2[2] = 1
li
```

```python
# in case of variable

def increment(a):
    a = a + 2
    return

a = 1
increment(a)
print(a)
```

```python
# in case of lists (list are mutable)

def increment(li):
    li[0] = li[0] + 2
    return
    
li = [1,2,3,4]
increment(li)
print(li)
```

```python
# in case of lists (list are mutable)

def increment(li):
    li = [7,7,8]
    return
    
li = [1,2,3,4]
increment(li)
print(li)

# but is return value is specified and is stored in a variable then the result would be different
```
