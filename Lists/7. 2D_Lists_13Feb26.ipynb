```python
# Row Sum
l = [[1,1,1], [2,2,2], [3,3,3]]
for row in range(len(l)):
    sum = 0
    for col in range(len(l[0])):
        sum += l[row][col]
    print('Sum of', row, 'th row is: ',sum)
print()
# Col Sum
l = [[1,1,1], [2,2,2], [3,3,3]]
for row in range(len(l)):
    sum = 0
    for col in range(len(l[0])):
        sum += l[col][row]
    print('Sum of', row, 'th col is: ',sum)
```

```python
# Transpose of a matrix or 90 degree rotation of a 2-D List

m = [[2,6,3], 
     [7,8,12], 
     [13,4,6]]
transpose = []
for row in range(len(m)):
    temp = []
    for col in range(len(m[0])):
        temp.append(m[col][row])
    transpose.append(temp)

for i in range(len(transpose)):
    for j in range(len(transpose[0])):
        print(transpose[i][j], end = " ")
    print()
```
