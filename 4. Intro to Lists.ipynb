```python
# Program to check if the input num is armstrong or not.

num = int(input())
temp = num
digitcount = len(str(num))
ans = 0
while (temp > 0):
    lastDigit = temp % 10   
    ans += lastDigit ** digitcount
    temp = temp // 10

if ans == num:
    print("The num is Armstrong")
else:
    print("Not Armstrong")


```
