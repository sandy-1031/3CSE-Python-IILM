```python
# Given a string, compute recursively a new string where all 'x' chars have been removed. xaxbx - ab

def removex(s):
    #base case
    if len(s) == 0:
        return s

    # Induction Hypothesis
    smallOutput = removex(s[1:])

    #induction Step
    if s[0] == 'x':
        return smallOutput
    else:
        return s[0] + smallOutput

removex("xaxbx")
```

```python
# Replacing the character a to b in s string recursively

def replace(s, a, b):
    # base case
    if len(s) == 0:
        return s

    # Induction Hypothesis
    smallOutput = replace(s[1:], a, b)

    #induction Step
    if s[0] == a:
        return b + smallOutput
    else:
        return s[0] + smallOutput

print(replace("abadagahaanzjnahjdhafljzmja", 'a', "z"))
```

```python
# Remove consecutive duplicates from string, xyyyzz - xyz

def removedup(s):
    if len(s) == 0 or len(s) == 1:
        return s
    smallOutput = removedup(s[1:])

    if s[0] == s[1]: 
        return smallOutput
    else:
        return s[0] + smallOutput
        

removedup("xxxyyzzzzzzzzz")
```

```python
# Replace 'pi' with 3.14 ; abcpi - abc3.14

def replacepi(s):
    if len(s) == 0 or len(s) == 1:
        return s
    #smallOutput = replacepi(s[1:])
    if s[0] == "p" and s[1] == 'i':
        return "3.14" + replacepi(s[2:])
    else:
        return s[0] + replacepi(s[1:])

replacepi("abcpi")
```
