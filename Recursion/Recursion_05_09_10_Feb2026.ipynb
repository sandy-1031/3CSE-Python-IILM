```python
# Replace 'pi' with 3.14 ; abcpi - abc3.14  --   piabcpjk

def replace_pi(s):
    #base case
    if len(s) <= 0:# and len(s) == 1:
        return s

    # induction hypothesis
    # 2 recursion calls
    if s[0] == "p" and s[1] =="i":
        return "3.14" + replace_pi(s[2:])
    else:
        return s[0] + replace_pi(s[1:])

replace_pi('abcpi')
```

```python
#Binary Search Recursive

def binaryRecursive(arr, k, si, ei):
    if si > ei:
        return -1

    mid = (si + ei) // 2
    if arr[mid] == k:
        return mid
    elif arr[mid] < k:
        return binaryRecursive(arr, k, mid + 1, ei)
    else:
        return binaryRecursive(arr, k, si, mid - 1)

ans = binaryRecursive([1,3,4,5,7,8,11,15,16,19,23], 15, 0, 10)
print(ans)
```

```python
def binarySearch(arr, ele, si, ei):
    if(si > ei):
        return -1
    
    mid = (si + ei)//2
    if(ele == arr[mid]):
        return mid
    elif(ele < arr[mid]):
        return binarySearch(arr, ele, si, mid - 1)
    else:
        return binarySearch(arr, ele, mid + 1, ei)
```

```python
ans = binarySearch([1,3,4,5,7,8,11,15,16,19,23], 15, 0, 10)
print(ans)
```

```python
def merge(a1, a2, a):
    i = 0
    j = 0
    k = 0
    while(i < len(a1) and j < len(a2)):
        if(a1[i] < a2[j]):
            a[k] = a1[i]
            k += 1
            i += 1
        else:
            a[k] = a2[j]
            k += 1
            j += 1
    while i < len(a1):
        a[k] = a1[i]
        k += 1
        i += 1
    while j < len(a2):
        a[k] = a2[j]
        k += 1
        j += 1
```

```python
def mergeSort(a):
    
    if(len(a) == 0 or len(a) == 1):
        return
    
    mid = len(a)//2
    a1 = a[0:mid]
    a2 = a[mid:]
    
    mergeSort(a1)
    mergeSort(a2)
    
    merge(a1, a2, a)
```

```python
def partition(a, si, ei):
    pivot = a[si]
    # find the number of elements smaller than pivot
    c = 0
    for i in range(si, ei + 1):
        if a[i] < pivot:
            c += 1
            
    a[si + c], a[si] = a[si], a[si + c]
    pivot_index = si + c
    
    while si < ei:
        if a[si] < pivot:
            si += 1
        elif a[ei] >= pivot:
            ei -= 1
        else:
            a[si], a[ei] = a[ei], a[si]
            si += 1
            ei -= 1
    return pivot_index
```

```python
def quickSort(a, si, ei):
    if(si >= ei):
        return
    
    pivot_index = partition(a, si, ei)
    quickSort(a, si, pivot_index - 1)
    quickSort(a, pivot_index + 1, ei)
```

```python
def tower_of_hanoi(n, source, aux, dest):
    # a = source, b = aux, c = dest
    #base case
    if n == 1:
        print('move 1st disk from', source, 'to', dest)
        return
    tower_of_hanoi(n-1, source, dest, aux)
    print('move', n, "th disk from", source, 'to', dest)
    tower_of_hanoi(n-1, aux, source, dest)

tower_of_hanoi(5, 'source', 'aux', 'dest')
```
