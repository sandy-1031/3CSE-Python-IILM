```python
#Check if list is sorted or not (using slicing + RECURSION)

def checkSorted(arr):
    #base case
    if len(arr) <= 1:
        return False

    # Ind Hypo
    
    #induction Step
    if arr[0] > arr[1]:
        return False
    ans = checkSorted(arr[1:])
    return ans

arr = [1,2,3,4,5,6,7,1]
checkSorted(arr)
```

```python
# Using start Index as reference

def isSortedBetter(a, si):          #si = starting index
    l = len(a)
    if(si == l - 1 or si == l):
        return True
    if(a[si] > a[si + 1]):
        return False
    return isSortedBetter(a, si + 1)

lis = [1,2,3,4,5, 7, 0]
isSortedBetter(lis, 0)
```
