```python
#Check if list is sorted or not (using slicing + RECURSION)

def checkSorted(arr):
    #base case
    if len(arr) <= 1:
        return False

    # Ind Hypo
    
    #induction Step
    if arr[0] > arr[1]:
        return False
    ans = checkSorted(arr[1:])
    return ans

arr = [1,2,3,4,5,6,7,1]
checkSorted(arr)
```

```python
# Using start Index as reference

def isSortedBetter(a, si):          #si = starting index
    l = len(a)
    if(si == l - 1 or si == l):
        return True
    if(a[si] > a[si + 1]):
        return False
    return isSortedBetter(a, si + 1)

lis = [1,2,3,4,5, 7, 0]
isSortedBetter(lis, 0)
```

```python
# CHECK if List is SORTED USING SLICING


def checkSorted(arr):
    l = len(arr)             #BASE CASE
    if (l ==0 or l == 1):
        return True
    
    smallerList = arr[1:]    #IH
    
    if arr[0] > arr[1]:      # CHECKING FOR 0th Element
        return False

    
    smallerListOutput = checkSorted(smallerList)      # IH - CALLING RECURSION ON REST OF THE LIST
    if smallerListOutput:
        return True
    else:
        return False
    
l = [1,2,3,4]
print(checkSorted(l))

```

```python
# USING INDEXING

def checkSortedBetter(arr, si):
    #BAse case
    if si == len(arr) or si == len(arr)-1:
        return True
    if arr[si] > arr[si+1]:
        return False

    smallOutput = checkSortedBetter(arr, si+1)
    return smallOutput
l = [1,2,3,4, 2]
print(checkSortedBetter(l, 0))
```
