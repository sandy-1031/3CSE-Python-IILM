```python
# Program to Print 1 to N using Recursion
def print1ton(n):
    #BASE CASE
    if n == 1:
        return 1

    #INDUCTION HYPOTHESIS
    smallOutput = print1ton(n-1)

    #Step 3: INDUCTION STEP
    print(n)
    return


n = int(input())
print1ton(n)
```

```python
# WAP to print n to 1 using recursion:
def printnto1(n):
    #Base Case
    if n == 0:
        return
    print(n)
    #Induction Hypothesis
    printnto1(n - 1)
    #Induction Step
    return

printnto1(10)
```

```python
#WAP to print the sum of first n natural numbers
def sumn(n):
    if n == 1:   #BASE CASE
        return 1
    smallOutput = sumn(n-1) #Induction Hypothesis
    return smallOutput + n

sumn(5)
```

```python
# WAP to find the sum of all elements of the array/list recursively

def sumlist(arr):
    if len(arr) == 1:
        return arr[0]
    smallOutput = sumlist(arr[:len(arr)-1])
    return smallOutput + arr[-1]

sumlist([5, 6, 9, 10, 5, 4, 2])
```
