```python
# Return the first index of integer x present in the array (use slicing)
def firstIndex(arr, x):
    #Base Case
    if len(arr) == 0:
        return -1

    #Check for 0th Index

    if arr[0] == x:
        return 0

    smallOutput = firstIndex(arr[1:], x)

    if smallOutput == -1:
        return -1
    else:
        return smallOutput + 1


a = [1,2,3,4,5,6,7,8,9,7]
x = 100
print(firstIndex(a, x))
```

```python
# Return the first index of integer x present in the array (without using slicing)

def firstIndexBetter(arr, x, si):
    #Base Case
    if len(arr) == si:
        return -1

    if arr[si] == x:
        return si

    smallOutput = firstIndexBetter(arr, x, si+1)
    
    return smallOutput

a = [1,2,3,4,5,6,7,8,9,7]
x = 10
print(firstIndexBetter(a, x, 0))
```

```python
# Return the last index of integer x present in the array (use slicing)

def lastIndex(arr, x):
    if len(arr) == 0:
        return -1

    if arr[len(arr)-1] == x:
        return len(arr) - 1

    smallOutput = lastIndex(arr[:len(arr)-1], x)

    if smallOutput == -1:
        return -1
    return smallOutput



a = [1,3,7,8,5,7,7,5,4,3,6,7,5,6]
x = 7
print(lastIndex(a, x))
```

```python
# Return the last index of integer x present in the array (use index reference)
def lastIndexBetter(arr, x, si):
    if si == len(arr):
        return -1

    smallOutput = lastIndexBetter(arr, x, si+1)

    if smallOutput != -1:
        return smallOutput 
    else:
        if arr[si] == x:
            return si
        return -1
        
a = [1,3,7,8,5,7,7,5,4,3,6,7,5,6]
x = 7
print(lastIndexBetter(a, x, 0))
```
