```python
# Program to find the HCF of 2 numbers - APPROACH 1
n1 = int(input())
n2 = int(input())
maxValue =   min(n1, n2)
ans = 0
for i in range(1, maxValue+1):
    if (n1%i == 0) and (n2 % i == 0) :
        ans = i

print(ans)
```
