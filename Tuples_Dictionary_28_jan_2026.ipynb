```python
# Program to print the fibonacci series till n

n = int(input())
num1 = 0
num2 = 1
print(num1, end = " ")
print(num2, end = " ")
s = 0
for i in range(3, n+1):
    s = num1 + num2
    print(s, end = " ")
    num1 = num2
    num2 = s
    
    #SWAP n1 = n2, n2 = s
```
