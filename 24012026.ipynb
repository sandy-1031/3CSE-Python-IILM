```python
# Basic structure of the code

n = int(input())

i = 0
while i < n:
    j = 0
    while j < n:
        print("*", end = " ")
        j += 1
    i += 1
    print()
```

```python
# Right Triangle Pattern

n = int(input())

i = 1
while i <= n:
    j = 1
    while j <= i:
        print(j, end = " ")
        j += 1
    i += 1
    print()
```

```python
# Right Triangle Pattern

n = int(input())

i = 1
while i <= n:
    j = 1
    while (j <= n - i + 1):
        print("*", end = " ")
        j += 1
    i += 1
    print()
```
