```python
n=4
spaces=n
p=1
for i in range(1,n+1):
    spaces=spaces-1
    p=i-1


    for j in range(1,spaces+1):
        print(" ",end=" ")
        
    for j in range(1,2*i):
        if(j>2*i/2):
            p-=1
        else:
            p+=1
        print(p,end=" ")
        
        
        
    print()
```
