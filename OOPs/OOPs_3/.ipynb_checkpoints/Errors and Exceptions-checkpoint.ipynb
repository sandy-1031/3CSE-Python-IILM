```python
# exception Handling

try :
    n = int(input("Enter the Numerator : "))
    d = int(input("Enter the Denominator : "))
    
    value = n / d
    print(value)
    
except ValueError :
    print("Numerator and Denominator should be Integer...")
```

```python
# if don't speify anything, then by default value of except is printed in case of any error
# specifying a particular error will only handle that error

while True :
    try :
        n = int(input("Enter the Numerator : "))
        d = int(input("Enter the Denominator : "))
        
        value = n / d
        print(value)
        break
    
    except :
        print("Numerator and Denominator should be Integer...")
```

## HANDING MULTIPLE EXCEPTIONS

```python


try :
    n = int(input("Enter the Numerator : "))
    d = int(input("Enter the Denominator : "))
    
    value = n / d
    print(value)
    
except ValueError :
    print("Numerator and Denominator should be Integer...")
    
except ZeroDivisionError :
    print("Denominator cannot be zero...")
    
```

```python
# or

try :
    n = int(input("Enter the Numerator : "))
    d = int(input("Enter the Denominator : "))
    
    value = n / d
    print(value)
    
except (ValueError, ZeroDivisionError) :
    print("Exception Occured...")
    
```
