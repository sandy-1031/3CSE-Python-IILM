### Polymorphism - ability to take multiple forms ( by object )
### Method Over-riding = when same funtion is there  -  the first call is for the function in child class and if we comment out that same function, then calling the function will be responded by the one in parent class

### in case you call self.print() - recursion will happena dn kernel will be dead

```python
class Vehicle:

    def __init__(self, color, maxSpeed):
        self.color = color
        self.__maxSpeed = maxSpeed
        
    def getMaxSpeed(self):
        return self.__maxSpeed
    
    def setMaxSpeed(self, maxSpeed):
        self.__maxSpeed = maxSpeed
        
    def print(self):
        print('Color =', self.color)
        print('maxSpeed =', self.__maxSpeed)
        
# Class Car will inherit the properties of Class Vehicle

class Car(Vehicle):
    def __init__(self, color, maxSpeed, numGears, isConvertible):
        super().__init__(color, maxSpeed)
        self.numGears = numGears
        self.isConvertible = isConvertible
        ''' 
    def print(self):
        #super().print()                                    #### or you can use  =  self.print() - a subclass inherits all properties and methods from super class
        print('numGears = ', self.numGears)
        print('isConvertible =', self.isConvertible)
'''
        # if in above function i use self.Print() to check the oreder of calling
        # then it will stuck in recursion because it will keep calling the self. print()
        # of the child class, hence prefer using super().print()
```
