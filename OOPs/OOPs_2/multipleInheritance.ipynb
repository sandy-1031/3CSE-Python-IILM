#### Multiple Base Classes which are inherited by single child class

```python
class Mother:
    
    def __init__(self, name):
        self.name = 'Mothers Name'

    def print(self):
        print('Print of Mother is called')
        
class Father:
    
    def __init__(self, name):
        self.name = 'Fathers Name'
        
    def print(self):
        print('Print of Father is called')
        
class Child(Mother, Father):
    
    def __init__(self, name):
        super().__init__(name)
        
    def printChild(self):
        print('Name of child is = ', self.name)
```

```python
c = Child("Amardeep")
c.printChild()
```

```python
d = Child('Amar')
c.print()
```

#### Hence, print() is called in the order of argument of the child class (of super classes)

```python
e = Child("abc")
e.print()
```
