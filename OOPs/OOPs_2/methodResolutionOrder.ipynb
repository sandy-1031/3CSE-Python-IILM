```python
class Mother:
    
    def __init__(self):
        self.name = 'Mothers Name'

    def print(self):
        print('Print of Mother is called')
        
class Father:
    
    def __init__(self):
        self.name = 'Fathers Name'
        
    def print(self):
        print('Print of Father is called')
        
class Child(Mother, Father):
    
    def __init__(self):
        super().__init__()
        
    def print(self):
        print('Name of child is = ', self.name)
```

```python
c = Child()
c.print()
print(Child.mro())
```

#### Hence, it gives the order of method resolution in a class

```python
class Mother:
    
    def __init__(self):
        self.name = 'Mothers Name'
        super().__init__()     ###### will go to __init__ of father class

    def print(self):
        print('Print of Mother is called')
        
class Father:
    
    def __init__(self):
        self.name = 'Fathers Name'
        super().__init__()                 #### Again executing it will do nothing
        
    def print(self):
        print('Print of Father is called')
        
class Child(Mother, Father):
    
    def __init__(self):
        super().__init__()
        
    def print(self):
        print('Name of child is = ', self.name)
```

```python
c = Child()
c.print()
print(Child.mro())
```
