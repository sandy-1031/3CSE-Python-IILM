#### Parent or Base or Super Class  -->  Child or Derived or Sub-Class
#### Derive class inherit property from Base class

```python
class Vehicle:
    def __init__(self, color, maxSpeed):
        self.color = color
        self.maxSpeed = maxSpeed
        
# Class Car will inherit the properties of Class Vehicle

class Car(Vehicle):
    def __init__(self, color, maxSpeed, numGears, isConvertible):
        super().__init__(color, maxSpeed)
        self.numGears = numGears
        self.isConvertible = isConvertible
        
    def printCar(self):
        print('Color = ', self.color)
        print('maxSpeed = ', self.maxSpeed)
        print('numGears = ', self.numGears)
        print('isConvertible =', self.isConvertible)
        
```

```python
c = Car('Yellow', 250, 7, True)
c.printCar()
```

## Inheritance and Private Members

```python
class Vehicle:
    def __init__(self, color, maxSpeed):
        self.color = color
        self.__maxSpeed = maxSpeed
        
    def getMaxSpeed(self):
        return self.__maxSpeed
    
    def setMaxSpeed(self, maxSpeed):
        self.__maxSpeed = maxSpeed
        
# Class Car will inherit the properties of Class Vehicle

class Car(Vehicle):
    def __init__(self, color, maxSpeed, numGears, isConvertible):
        super().__init__(color, maxSpeed)
        self.numGears = numGears
        self.isConvertible = isConvertible
        
    def printCar(self):
        print('Color = ', self.color)
        print('maxSpeed = ', self.getMaxSpeed())
        print('numGears = ', self.numGears)
        print('isConvertible =', self.isConvertible)
        
```

```python
c = Car('Yellow', 250, 7, True)
c.printCar()
```

## More on Inheritance

```python
class Vehicle:
    def __init__(self, color, maxSpeed):
        self.color = color
        self.__maxSpeed = maxSpeed
        
    def getMaxSpeed(self):
        return self.__maxSpeed
    
    def setMaxSpeed(self, maxSpeed):
        self.__maxSpeed = maxSpeed
        
    def print(self):
        print('Color =', self.color)
        print('maxSpeed =', self.__maxSpeed)
        
# Class Car will inherit the properties of Class Vehicle

class Car(Vehicle):
    def __init__(self, color, maxSpeed, numGears, isConvertible):
        super().__init__(color, maxSpeed)
        self.numGears = numGears
        self.isConvertible = isConvertible
        
    def printCar(self):
        super().print()                                    #### or you can use  =  self.print() - a subclass inherits all properties and methods from super class
        print('numGears = ', self.numGears)
        print('isConvertible =', self.isConvertible)
        
```

```python
c = Car('Orange', 270, 8, True)
c.printCar()
```
