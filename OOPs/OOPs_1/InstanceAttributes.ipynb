#### Class - Is a template, which tells what kind of data and functionality objects shoud have.

#### Object - is instance of this class

##### Eg: When FRUIT is a class, its instance will be APPLE, ORANGE, MANGO - which is the object

```python
class Student:
    pass

# Empty class
```

```python
s1 = Student()
```

```python
print(type(s1))
```

```python
s2 = Student()
s3 = Student()
type(s2)
```

```python
print(s2, s3)
```

### Assigning values to Object

```python
s1.name = 'Sandeep'
s2.name = "Student2"

# here name is the attribute
```

```python
s2.rollNumber = 22
s1.rollNumber = 21
```

```python
s3.name = 'Student3'
s3.rollNumber = 23
```

```python
s3.rollNumber 
print(s2.rollNumber)
```

```python
## to see all attributes of object

print(s3.__dict__, s2.__dict__, s1.__dict__)
```

```python
s2.__dict__
```

```python
s1.__dict__
```

```python
# has attribute in object

hasattr(s1, 'name')
```

```python
# get attribut in object
# 3rd attributes is returned when the 2nd one in object is not present   (Default Value)

getattr(s1, 'name')
```

```python
getattr(s1, 'rollNumber', '')
```

```python
# delete attribute

delattr(s3, 'name')
```
