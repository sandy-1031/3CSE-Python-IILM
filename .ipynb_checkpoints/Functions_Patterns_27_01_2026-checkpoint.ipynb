```python
# Pattern questions
n = 4
i = 1
while i <= n:
    j = 1
    #Printing empty space
    while j <= n-i:
        print(" ", end = " ")
        j += 1
    # Printing increasing numbers starting from i to 2*i
    k = i
    while k < (2*i):
        print(k, end = " ")
        k += 1
    # Incrementing i and going to new line
    i += 1
    print()
```

```python
#Function: to implement modularity or to resuse a set of code
def sumOfNums(a, b, c):
    s = a + b + c
    return(s)

ans = sumOfNums(1,2,3)
print(ans)   
```

```python
# SCOPE OF VARIABLES

def scope():
    #LOCAL VARIABLE
    b = 10
    print(b)
    return

# GLOBAL VARIABLE
a = 20
scope()
print(a)
print(b)
```
